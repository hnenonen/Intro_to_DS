```python
import pandas as pd
import nltk
```

overview : the original overview
overview_punct_sep: The overview as a list, with end- and beginning- of word punctuation separated
overview_lower: lowercase, with no punctuation


```python
data_older = pd.read_csv('data_1980to2000.csv')
data_newer = pd.read_csv('data_2000to2023.csv')
data = pd.concat(objs=[data_older, data_newer])
```

```python
# separates punctuation in end and beginning of word from the words themselves
# punctuation inside words is kept as-is
data['overview_punct_sep'] = data['overview']
data.overview_punct_sep = data.overview_punct_sep.str.replace('^(\W+)(\w)', r'\1 \2', regex=True)
data.overview_punct_sep = data.overview_punct_sep.str.replace('(\w)(\W+)$', r'\1 \2', regex=True)
data.overview_punct_sep = data.overview_punct_sep.str.replace('(\w)(\W+)(\s+)', r'\1 \2\3', regex=True)
data.overview_punct_sep = data.overview_punct_sep.str.replace('(\s+)(\W+)(\w)', r'\1\2 \3', regex=True)

#split into words
data.overview_punct_sep = data.overview_punct_sep.str.split()

from nltk import word_tokenize
nltk.download('punkt')
data['overview_tokenize'] = data.overview.apply(lambda x: nltk.word_tokenize(x))
data['overview_sent_tokenize'] = data.overview.apply(lambda x: nltk.sent_tokenize(x))
data['overview_sent_tokenize'] = data.overview_sent_tokenize.apply(lambda x: [nltk.word_tokenize(sent) for sent in x])


with pd.option_context('display.max_colwidth', 1000):
  display(data)
```

```python
# overview_lower is a version of the overview lowercase and nnon-mid-word-punctuation removed

data['overview_lower'] = data.overview_tokenize

data.overview_lower = data.overview_lower.apply(lambda x: [word.lower() for word in x])

with pd.option_context('display.max_colwidth', 1000):
  display(data)
```

```python
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')

data['overview_stopwords'] = data['overview_lower']
data.overview_stopwords = data.overview_stopwords.apply(lambda x: [word for word in x if word not in stopwords])

with pd.option_context('display.max_colwidth', 1000):
  display(data)
```

```python
data['overview_stemmed'] = data.overview_lower

from nltk.stem.porter import *
ps = PorterStemmer()
data.overview_stemmed = data.overview_stemmed.apply(lambda x: [ps.stem(word) for word in x])

with pd.option_context('display.max_colwidth', 1000):
  display(data)
```

```python
data['overview_pos'] = data.overview_lower
nltk.download('averaged_perceptron_tagger')
data.overview_pos = data.overview_pos.apply(lambda x: nltk.pos_tag(x))


with pd.option_context('display.max_colwidth', 1000):
  display(data)
```
